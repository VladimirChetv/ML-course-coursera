```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

# Распределение арксинуса

Плотность вероятности: $ f(x) = \frac{1}{\pi \sqrt{x(1-x)}} $

Функция распределения: $ F(x) = \frac{2}{\pi}\arcsin(\sqrt{x}) $

Математическое ожидание: $ \frac{1}{2}$

Дисперсия:  $ \frac{1}{8}$


### Сгенирируем выборку объёма 1000, построим гистограмму выборки и нарисуем поверх неё теоретическую плотность распределения

```python
arcsin = sts.arcsine()
x = arcsin.rvs(size=1000)
plt.hist(x, 50, density=True)

t = np.arange(0.003, 0.998, 0.001)
f = [1/(math.pi * (i * (1 - i)) ** 0.5) for i in t]

plt.plot(t, f)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

#### Сгенерируйте 1000 выборок объёма n

```python
def foo(n, var):
    x = [np.mean((arcsin.rvs(size=n))) for i in range(1000)]
    plt.hist(x, 70, density=True)

    t = np.arange(0.001, 0.999, 0.001)

    plt.plot(t, sts.norm(loc=0.5, scale=var ** 0.5).pdf(t))
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
```

####  Построим гистограммы распределений их выборочных средних для выборок объема 5, 16, 60. А также определим параметры нормального распределения $ Ex = \frac{1}{2} $ и $ Dx = \frac{1}{8n} $

```python
n = 5
varience = 1 / (8 * n)
foo(n, varience)
```

```python
n = 10
varience = 1 / (8 * n)
foo(n, varience)
```
